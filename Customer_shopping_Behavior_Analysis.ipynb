```python
# Checking the missing value
df.isnull().sum()
```

```python
# Replacing null value (NaN) in the column Review Rating with the median rating of the same category
# Keep missing data handling category-specific (more accurate than using a global median.

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# formatting the column name with lowercase and underscore. Easy to read and easy to reference in SQL 
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Creating a column age_group: Very useful to analyze across different age groups to undersand the customer behaviour
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Create column purchase_frequency_days : Convert the text into days and makes the analysis very easy
frequency_mapping = {
    'Fortnightly': 14,     # also fixed spelling from "Forthnightly" → "Fortnightly"
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
